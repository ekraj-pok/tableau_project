```python
import pandas as pd
import numpy as np
```

```python
goods_export_df = pd.read_excel("export_goods_country.xlsx")
```

```python
service_export_df.rename(columns = {'Exporters' : 'country', 'Exported value in 2003': "2003", 'Exported value in 2004':'2004',
       'Exported value in 2005': '2005', 'Exported value in 2006': '2006',
       'Exported value in 2007': '2007','Exported value in 2008': '2008',
       'Exported value in 2009': '2009', 'Exported value in 2010': '2010',
       'Exported value in 2011':'2011', 'Exported value in 2012':'2012',
       'Exported value in 2013': '2013', 'Exported value in 2014': '2014',
       'Exported value in 2015': '2015', 'Exported value in 2016':'2016',
       'Exported value in 2017':'2017', 'Exported value in 2018':'2018',
       'Exported value in 2019': '2019', 'Exported value in 2020':'2020',
       'Exported value in 2021':'2021'}, inplace = True)

columns_to_divide = [ "2003", '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021']
service_export_df[columns_to_divide] = service_export_df[columns_to_divide]/(1000000)
service_export_df = service_export_df.fillna(0)
final_export_service = pd.melt(service_export_df, id_vars= ["country"], var_name="Year")

final_export_service  = final_export_service.set_index(["country", "Year"])
final_export_service.rename(columns = {"value": "service_export"}, inplace = True)
final_export_service
```

```python
goods_import_df = pd.read_excel("import_goods_country.xlsx")
goods_import_df.drop(columns = ['Imported value in 2022'], inplace = True)
```

```python
goods_import_df.rename(columns = {'Importers' : 'country', 'Imported value in 2003': "2003", 'Imported value in 2004':'2004',
       'Imported value in 2005': '2005', 'Imported value in 2006': '2006',
       'Imported value in 2007': '2007','Imported value in 2008': '2008',
       'Imported value in 2009': '2009', 'Imported value in 2010': '2010',
       'Imported value in 2011':'2011', 'Imported value in 2012':'2012',
       'Imported value in 2013': '2013', 'Imported value in 2014': '2014',
       'Imported value in 2015': '2015', 'Imported value in 2016':'2016',
       'Imported value in 2017':'2017', 'Imported value in 2018':'2018',
       'Imported value in 2019': '2019', 'Imported value in 2020':'2020',
       'Imported value in 2021':'2021'}, inplace = True)

columns_to_divide = [ "2003", '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021']
goods_import_df[columns_to_divide] = goods_import_df[columns_to_divide]/1000000
goods_import_df = goods_import_df.fillna(0)
final_import_goods = pd.melt(goods_import_df, id_vars= ["country"], var_name="Year")

final_import_goods = final_import_goods[final_import_goods["country"] != "World"]

final_import_goods = final_import_goods.set_index(["country", "Year"])
final_import_goods.rename(columns = {"value": "goods_import"}, inplace = True)

final_import_goods
```

```python
service_import_df = pd.read_excel("import_service_country.xlsx")
service_import_df.drop(columns = ['Imported value in 2002'], inplace = True)
```

```python
service_import_df.rename(columns = {'Importers' : 'country', 'Imported value in 2003': "2003", 'Imported value in 2004':'2004',
       'Imported value in 2005': '2005', 'Imported value in 2006': '2006',
       'Imported value in 2007': '2007','Imported value in 2008': '2008',
       'Imported value in 2009': '2009', 'Imported value in 2010': '2010',
       'Imported value in 2011':'2011', 'Imported value in 2012':'2012',
       'Imported value in 2013': '2013', 'Imported value in 2014': '2014',
       'Imported value in 2015': '2015', 'Imported value in 2016':'2016',
       'Imported value in 2017':'2017', 'Imported value in 2018':'2018',
       'Imported value in 2019': '2019', 'Imported value in 2020':'2020',
       'Imported value in 2021':'2021'}, inplace = True)

columns_to_divide = [ "2003", '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021']
service_import_df[columns_to_divide] = service_import_df[columns_to_divide]/(1000000)
service_import_df = service_import_df.fillna(0)
final_import_service = pd.melt(service_import_df, id_vars= ["country"], var_name="Year")
final_import_service.rename(columns = {"value": "service_import"}, inplace = True)
final_import_service = final_import_service.set_index(["country", "Year"])

final_import_service
```

```python
import_df = final_import_service.join(final_import_goods, how = "outer")
export_df = final_export_service.join(final_export_goods, how = "outer")
```

```python
export_df["total_export"] = export_df["service_export"] + export_df["goods_export"]
export_df
```

```python
import_df["total_import"] = import_df["service_import"] + import_df["goods_import"]

import_df.info()
```

```python
trade_df = import_df.join(export_df, how = "outer").reset_index()
trade_df["trade_surplus"] = trade_df["total_export"] - trade_df["total_import"]
```

```python
trade_df.to_excel("final_trade_data.xlsx", index = False)
```
